## Subplots

```python
df1 = pd.DataFrame(columns = ['data'], data=np.random.randn(1000)*50)

# Plotting Train Data
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(20, 15), num='Subplots')

# Figure 1
g1 = sns.scatterplot(x=df1.index, y=df1.data, ax=ax1)
ax1.set_ylabel('data')
ax1.grid(True)
ax1.set_title('Train Data')
ax1.set_ylim(-170, 170)
ax1.set(xlabel=None)

# Figure 2
df1['prediction'] = ((df1.data>-100) & (df1.data<100)).astype(int) 
legend_map = {0: 'Anomaly', 1: 'Normal'}
palette = {'Anomaly': 'red', 'Normal': 'blue'}
g2 = sns.scatterplot(x=df1.index, y=df1.data, hue=df1.prediction.map(legend_map), palette=palette, ax=ax2, legend='full')
ax2.grid(True)
ax2.set_ylabel('data')
ax2.set_title('Tested on Locally learned Model')
ax2.set_ylim(-170, 170)
ax2.set(xlabel=None)

# Figure 3
df1['prediction'] = ((df1.data>-80) & (df1.data<80)).astype(int) 
g3 = sns.scatterplot(x=df1.index, y=df1.data, hue=df1.prediction.map(legend_map), palette=palette, ax=ax3, legend='full')
ax3.grid(True)
ax3.set_title('Tested on Global learning Model')
ax3.set_ylim(-170, 170)
ax3.set_ylabel('data')
ax3.set_xlabel('time')
plt.show()
```
