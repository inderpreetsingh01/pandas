```python
def func(s):
    r = s<2000
    return ["background-color:red" if i else "background-color:blue" for i in r]
```

```python
def func1(s):
    if 'Year' in s.name:
        r = s<2005
        return ["background-color:red" if i else "background-color:blue" for i in r] 
    else:
        return ["background-color:None" for i in s]
```

```python
df.style.apply(func1)
```
